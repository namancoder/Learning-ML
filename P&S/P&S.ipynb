```python
import numpy as np
import pylab
import scipy.stats as stats

#N(0,1)
# loc = mean
# scale = standard deviation

std_normal = np.random.normal(loc =0,scale =1,size=100)#generate random observation from gaussian dist.

#0 to 100th percentile
for i in range(0,101):
    print(i,np.percentile(std_normal,i))
```

```python
measurements = np.random.normal(loc=20,scale=5,size=100)

stats.probplot(measurements, dist='norm', plot = pylab)

#dist='norm' means standard normal viariate N(0,1)
pylab.show()
```

```python
from scipy import stats
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(121)
x = stats.loggamma.rvs(5,size=500) +5
# Q-Q plot agains X= log-gamma and Y=Normal
prob = stats.probplot(x,dist=stats.norm,plot=ax1)
ax1.set_xlabel('')
ax1.set_title('probplot against normal distribution')
ax2 = fig.add_subplot(122)
# X= Normal , Y= boxcox(X)
xt, _ =stats.boxcox(x)
prob = stats.probplot(xt,dist=stats.norm, plot=ax2)
ax2.set_title("BOX COX")
```
